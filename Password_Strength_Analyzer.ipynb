```python
from zxcvbn import zxcvbn
from getpass import getpass
import getpass
```

```python
def analyze_password(password, user_inputs=[]):
    if not password:
        print("No password entered")
        return

    result = zxcvbn(password, user_inputs=user_inputs)

    print("="*50)
    print("PASSWORD STRENGTH ANALYSIS")
    print("="*50)
    print("Score (0-4):", result['score'])
    print("Crack Time:",
          result['crack_times_display']['offline_slow_hashing_1e4_per_second'])
    print("Guesses:", result['guesses'])

    if result['feedback']['warning']:
        print("Warning:", result['feedback']['warning'])

    for s in result['feedback']['suggestions']:
        print("Suggestion:", s)
```

```python
pwd = getpass.getpass("Enter password: ")
user_data = ["Kashif", "1999"]

analyze_password(pwd, user_data)
```

```python
LEET_MAP = {
    'a': ['4', '@'],
    'e': ['3'],
    'i': ['1', '!'],
    'o': ['0'],
    's': ['5', '$']
}
```

```python
def apply_leetspeak(word):
    variations = {word}
    for char, subs in LEET_MAP.items():
        new_vars = set()
        for v in variations:
            if char in v:
                for s in subs:
                    new_vars.add(v.replace(char, s))
        variations |= new_vars
    return variations
```

```python
def generate_wordlist(inputs):
    wordlist = set()
    years = ['2023', '2024', '2025', '123']

    for word in inputs:
        for v in apply_leetspeak(word.lower()):
            wordlist.add(v)
            for y in years:
                wordlist.add(v + y)
    return wordlist
```

```python
inputs = ["Kashif", "Rocky", "1999"]
wordlist = generate_wordlist(inputs)

print("Total words generated:", len(wordlist))
print(list(wordlist)[:10])  # show sample
```
